### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Load dataset

```python
df = pd.read_csv('data/ai4i2020.csv')
```

### Exploratory Data Analysis

```python
df.info()
```

- There is no missing values

```python
df.head()
```

```python
df.describe()
```

- No negative values

```python
failure_rate = 100*(df['Machine failure'] == 1).sum() / df.shape[0]
non_failure_rate = 100*(df['Machine failure'] == 0).sum() / df.shape[0]
print('Failure Percentage: ', failure_rate, '%')
```

```python
sns.set_theme(style="darkgrid")
plt.bar(['Non-Failure', 'Failure'], [non_failure_rate/100, failure_rate/100]);
plt.xlabel('Non-Failure vs Failure')
plt.ylabel('Percentage')
plt.title('The Percentage of Failure');
```

- Highly imbalance dataset as expected
- Need to stratify our data, and use cross validation techniques when training the model

```python
sns.pairplot(df.iloc[:, 3:-5], hue='Machine failure', diag_kind="hist");
```

- **Rotational speed** is skewed to the right.
- **Air temperature** seems to be bimodel.
- **Tool wear** is mostly uniform.

- The failure mostly happens at the "edge" of the certain distribution.
- It seems that temperatures are not good indicators.
